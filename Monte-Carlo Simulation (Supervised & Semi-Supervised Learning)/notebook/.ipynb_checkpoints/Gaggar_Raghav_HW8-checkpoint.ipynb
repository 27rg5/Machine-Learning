### Question 1. b) i)

```python
df = df.iloc[:, 1:]
m_df = df.loc[df[1] == 1]
b_df = df.loc[df[1] == 0]
```

```python
def generate_dataset(data1, data2):
    M = 30
    montecarlo_random = random.randint(1,M)
    m_train_X, m_test_X, m_train_y, m_test_y = train_test_split(data1.iloc[:, 1:], data1.iloc[:, 0], test_size = 0.2, random_state = montecarlo_random)
    b_train_X, b_test_X, b_train_y, b_test_y = train_test_split(data2.iloc[:, 1:], data2.iloc[:, 0], test_size = 0.2, random_state = montecarlo_random)
    
    Xtrain = pd.DataFrame()
    ytrain = pd.DataFrame()
    Xtest = pd.DataFrame()
    ytest = pd.DataFrame()
    
    Xtrain = pd.concat([m_train_X, b_train_X])
    Xtest = pd.concat([m_test_X, b_test_X])
    ytrain = pd.concat([m_train_y, b_train_y])
    ytest = pd.concat([m_test_y, b_test_y])
    
    
    return Xtrain, Xtest, ytrain, ytest
```

```python
m_df1 = m_df.copy()
b_df1 = b_df.copy()

for column in m_df1.columns[1:]:
    m_df1[column] = (m_df1[column] - m_df1[column].min()) / (m_df1[column].max() - m_df1[column].min())
    b_df1[column] = (b_df1[column] - b_df1[column].min()) / (b_df1[column].max() - b_df1[column].min())
```

```python
accuracy_train = []
accuracy_test = []
precision_train = []
precision_test = []
recall_train = []
recall_test = []
f1_train = []
f1_test = []
auc_train = []
auc_test = []

for i in range (0, 30):
    train_X, test_X, train_y, test_y = generate_dataset(m_df1, b_df1)
    params = {'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000, 10000, 100000, 1000000]}
    SVM = LinearSVC(penalty='l1', dual=False)
    gridCV = GridSearchCV(SVM, params, cv=5).fit(train_X, train_y)
    print("\nIteration:", i+1)
    optimal_C = gridCV.best_params_['C']
    print('Most optimal C:', optimal_C)
    
    train_pred = gridCV.predict(train_X)
    test_pred = gridCV.predict(test_X)
    
    accuracy_train.append(accuracy_score(train_y, train_pred))
    accuracy_test.append(accuracy_score(test_y, test_pred))
    
    train_confusion_matrix = confusion_matrix(train_y, train_pred)
    test_confusion_matrix = confusion_matrix(test_y, test_pred)
    tn1, fp1, fn1, tp1 = train_confusion_matrix.ravel()
    tn2, fp2, fn2, tp2 = test_confusion_matrix.ravel()
    
    precision_train.append(tp1/(tp1 + fp1)) 
    precision_test.append(tp2/(tp2 + fp2))

    recall_train.append(tp1/(tp1 + fn1))
    recall_test.append(tp2/(tp2 + fn2))
    
    f1_train.append(f1_score(train_y, train_pred))
    f1_test.append(f1_score(test_y, test_pred))

    pred_prob_train = gridCV.decision_function(train_X)
    pred_prob_test = gridCV.decision_function(test_X)
    
    auc_train.append(roc_auc_score(train_y, pred_prob_train))
    auc_test.append(roc_auc_score(test_y, pred_prob_test))
```

```python
c_matrix = confusion_matrix(train_y, train_pred)
print("The confusion matrix for interation 30 is:\n", c_matrix)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = c_matrix, display_labels = [False, True])
cm_display.plot()
plt.show()
```

```python
c_matrix = confusion_matrix(test_y, test_pred)
print("The confusion matrix for interation 30 is:\n", c_matrix)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = c_matrix, display_labels = [False, True])
cm_display.plot()
plt.show()
```

```python
fpr1, tpr1, _ = roc_curve(train_y, pred_prob_train)
fpr2, tpr2, _ = roc_curve(test_y, pred_prob_test)
plt.plot(fpr1, tpr1, label='ROC curve (Train)')
plt.plot(fpr2, tpr2, label='ROC curve (Test)')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show() 
```

```python
avg_train_acc = sum(accuracy_train)/len(accuracy_train)
avg_test_acc = sum(accuracy_test)/len(accuracy_test)
avg_train_recall = sum(recall_train)/len(recall_train)
avg_test_recall = sum(recall_test)/len(recall_test)
avg_train_precision = sum(precision_train)/len(precision_train)
avg_test_precision = sum(precision_test)/len(precision_test)
avg_train_f1score = sum(f1_train)/len(f1_train)
avg_test_f1score = sum(f1_test)/len(f1_test)
avg_train_auc = sum(auc_train)/len(auc_train)
avg_test_auc = sum(auc_test)/len(auc_test)

print("\nAverage Train accuracy:", avg_train_acc)
print("Average Test accuracy:", avg_test_acc)
print("\nAverage Average Recall (Train):", avg_train_recall)
print("Average Recall (Test):", avg_test_recall)
print("\nAverage Precision (Train):", avg_train_precision)
print("Average Precision (Test):", avg_train_precision)
print("\nAverage F1 score (Train):", avg_train_f1score)
print("Average F1 score (Test):", avg_test_f1score)
print("\nAverage AUC score (Train):", avg_train_auc)
print("Average AUC score (Test):", avg_test_auc)
```

### Question 1. b) ii)

```python
accuracy_train = []
accuracy_test = []
precision_train = []
precision_test = []
recall_train = []
recall_test = []
f1_train = []
f1_test = []
auc_train = []
auc_test = []
for i in range (0, 30):
    train_X, test_X, train_y, test_y = generate_dataset(m_df1, b_df1)
    labelled_Xtrain = train_X.iloc[:train_X.shape[0]//2, :]
    labelled_ytrain = train_y.iloc[:train_X.shape[0]//2]
    new_Xtrain = labelled_Xtrain
    new_ytrain = labelled_ytrain
    params = {'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000, 10000, 100000, 1000000]}
    SVM = LinearSVC(penalty='l1', dual=False)
    gridCV = GridSearchCV(SVM, params, cv=5).fit(labelled_Xtrain, labelled_ytrain)
    print("\nIteration:", i+1)
    optimal_C = gridCV.best_params_['C']
    print('Most optimal C:', optimal_C)
    
    unlabelled_Xtrain = train_X.iloc[train_X.shape[0]//2:, :]
    unlabelled_ytrain = train_y.iloc[train_X.shape[0]//2:]
    while unlabelled_Xtrain.shape[0] > 0:
        boundary = gridCV.decision_function(unlabelled_Xtrain)
        dist = np.abs(boundary)
        Xfarthest = dist.argmax()
        if boundary[Xfarthest] < 0:
            yfarthest = 0
        else:
            yfarthest = 1
        new_Xtrain = np.append(new_Xtrain, [unlabelled_Xtrain.iloc[Xfarthest]], axis = 0)
        new_ytrain = new_ytrain.append(pd.Series(yfarthest))
        unlabelled_Xtrain = unlabelled_Xtrain.drop(unlabelled_Xtrain.index[Xfarthest])        
        SVM = LinearSVC(penalty='l1', dual=False, C=optimal_C)
        SVM.fit(new_Xtrain, new_ytrain)
        
    train_pred = SVM.predict(new_Xtrain)
    test_pred = SVM.predict(test_X)
    
    accuracy_train.append(accuracy_score(new_ytrain, train_pred))
    accuracy_test.append(accuracy_score(test_y, test_pred))
    
    train_confusion_matrix = confusion_matrix(new_ytrain, train_pred)
    test_confusion_matrix = confusion_matrix(test_y, test_pred)
    tn1, fp1, fn1, tp1 = train_confusion_matrix.ravel()
    tn2, fp2, fn2, tp2 = test_confusion_matrix.ravel()
    
    precision_train.append(tp1/(tp1 + fp1)) 
    precision_test.append(tp2/(tp2 + fp2))

    recall_train.append(tp1/(tp1 + fn1))
    recall_test.append(tp2/(tp2 + fn2))
    
    f1_train.append(f1_score(new_ytrain, train_pred))
    f1_test.append(f1_score(test_y, test_pred))

    pred_prob_train = SVM.decision_function(new_Xtrain)
    pred_prob_test = SVM.decision_function(test_X)
    
    auc_train.append(roc_auc_score(new_ytrain, pred_prob_train))
    auc_test.append(roc_auc_score(test_y, pred_prob_test))
```

```python
c_matrix = confusion_matrix(new_ytrain, train_pred)
print("The confusion matrix for interation 30 is:\n", c_matrix)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = c_matrix, display_labels = [False, True])
cm_display.plot()
plt.show()
```

```python
c_matrix = confusion_matrix(test_y, test_pred)
print("The confusion matrix for interation 30 is:\n", c_matrix)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = c_matrix, display_labels = [False, True])
cm_display.plot()
plt.show()
```

```python
fpr1, tpr1, _ = roc_curve(new_ytrain, pred_prob_train)
fpr2, tpr2, _ = roc_curve(test_y, pred_prob_test)
plt.plot(fpr1, tpr1, label='ROC curve (Train)')
plt.plot(fpr2, tpr2, label='ROC curve (Test)')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show() 
```

```python
avg_train_acc = sum(accuracy_train)/len(accuracy_train)
avg_test_acc = sum(accuracy_test)/len(accuracy_test)
avg_train_recall = sum(recall_train)/len(recall_train)
avg_test_recall = sum(recall_test)/len(recall_test)
avg_train_precision = sum(precision_train)/len(precision_train)
avg_test_precision = sum(precision_test)/len(precision_test)
avg_train_f1score = sum(f1_train)/len(f1_train)
avg_test_f1score = sum(f1_test)/len(f1_test)
avg_train_auc = sum(auc_train)/len(auc_train)
avg_test_auc = sum(auc_test)/len(auc_test)

print("\nAverage Train accuracy:", avg_train_acc)
print("Average Test accuracy:", avg_test_acc)
print("\nAverage Average Recall (Train):", avg_train_recall)
print("Average Recall (Test):", avg_test_recall)
print("\nAverage Precision (Train):", avg_train_precision)
print("Average Precision (Test):", avg_train_precision)
print("\nAverage F1 score (Train):", avg_train_f1score)
print("Average F1 score (Test):", avg_test_f1score)
print("\nAverage AUC score (Train):", avg_train_auc)
print("Average AUC score (Test):", avg_test_auc)
```
