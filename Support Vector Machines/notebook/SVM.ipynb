```python
df = pd.read_csv('../data/Frogs_MFCCs.csv')
df
```

```python
train_set = df.sample(frac = 0.7)
test_set = df.drop(train_set.index)
train_set
```

```python
train_X = train_set.iloc[:, :-4]
train_y = train_set.iloc[:, -4:]
test_X = test_set.iloc[:, :-4]
test_y = test_set.iloc[:, -4:]
```

```python
c_val = [0.001, 0.01, 0.1, 1, 10, 100, 1000, 10000, 100000, 1000000]
gamma_val = np.arange(0.1, 2.1, 0.1)

params = {"C":c_val, "gamma": gamma_val}

Family_SVM = GridSearchCV(SVC(kernel= "rbf", decision_function_shape = "ovr"), params, cv=10).fit(train_X, train_y['Family'])
Genus_SVM = GridSearchCV(SVC(kernel= "rbf", decision_function_shape = "ovr"), params, cv=10).fit(train_X, train_y['Genus'])
Species_SVM = GridSearchCV(SVC(kernel= "rbf", decision_function_shape = "ovr"), params, cv=10).fit(train_X, train_y['Species'])
```

```python
family_c = Family_SVM.best_params_['C']
family_gamma = Family_SVM.best_params_['gamma']
print("Family:")
print("Most optimal C:", family_c)
print("Most optimal Gamma:", family_gamma)
print()

genus_c = Genus_SVM.best_params_['C']
genus_gamma = Genus_SVM.best_params_['gamma']
print("Genus:")
print("Most optimal C:", genus_c)
print("Most optimal Gamma:", genus_gamma)
print()

species_c = Species_SVM.best_params_['C']
species_gamma = Species_SVM.best_params_['gamma']
print("Species:")
print("Most optimal C:", species_c)
print("Most optimal Gamma:", species_gamma)
```

```python
Family_SVM = SVC(C = family_c, gamma = family_gamma, kernel='rbf', decision_function_shape = "ovr").fit(train_X, train_y['Family'])
predictions = Family_SVM.predict(test_X)
print("Family:\n")
print("Most optimal C:", family_c)  
print("Most optimal Gamma:", family_gamma)
loss1 = hamming_loss(test_y['Family'], predictions)
print("Hamming Loss:", loss1)
score1 = 1 - loss1
print("Hamming Score:", score1)
```

#### The SVM with Gaussian kernel gives us the best results, as we get the highest hamming score and the least hamming loss through that model.

```python
def get_results(predictions, real):
    loss = hamming_loss(predictions, real)
    score = 1-loss
    return score, loss
```

```python
X = df.iloc[:, :22]
y = df.iloc[:, -4:]
ham_loss_final = []
ham_score_final = []
ham_dist_final = []
optimal = []
majority_label = []
best_k = 0
max_score = 0

for i in range(0, 50):
    score = {}
    montecarlo_random = random.randint(1,1000)
    for k in range(2, 51):
        model = KMeans(n_clusters=k, random_state = montecarlo_random).fit(X)
        labels = model.labels_
        score = silhouette_score(X, labels)
        if score > max_score:
            max_score = score
            best_k = k
    print("For iteration:", i)
    print('\nBest k:', best_k)
    print('Maximum silhouette score:', max_score)

    model = KMeans(n_clusters = best_k)
    model.fit(X)
    index = model.labels_
    final_label = []
    for k in range(0, best_k):
        label_genus = {}    
        label_species = {}
        label_family = {}
        for j in range(0, len(y)):
            if index[j] == k:
                real_species = y['Species'][j]
                if real_species in label_species.keys():
                    label_species[real_species]+=1
                else:
                    label_species[real_species] = 1
                real_genus = y['Genus'][j]
                if real_genus in label_genus.keys():
                    label_genus[real_genus]+=1
                else:
                    label_genus[real_genus] = 1
                real_family = y['Family'][j]
                if real_family in label_family.keys():
                    label_family[real_family]+=1
                else:
                    label_family[real_family] = 1

        family_cluster = sorted(label_family.items(), key = operator.itemgetter(1), reverse=True)[0][0]
        genus_cluster = sorted(label_genus.items(), key = operator.itemgetter(1), reverse=True)[0][0]
        species_cluster = sorted(label_species.items(), key = operator.itemgetter(1), reverse=True)[0][0]
        temp = {}
        temp['Family'] = family_cluster
        temp['Genus'] = genus_cluster
        temp['Species'] = species_cluster
        final_label.append(temp)
        

    pred_family = []
    pred_genus = []
    pred_species = []
    for j in range(0, len(X)):
        x = index[j]
        labels_ = final_label[x]
        family_lbl = labels_['Family']
        genus_lbl = labels_['Genus']
        species_lbl = labels_['Species']
        pred_family.append(family_lbl)
        pred_genus.append(genus_lbl)
        pred_species.append(species_lbl)

    real_family, pred_family = np.array(y['Family']), np.array(pred_family)
    real_genus, pred_genus = np.array(y['Genus']), np.array(pred_genus)
    real_species, pred_species = np.array(y['Species']), np.array(pred_species)
    family_score, family_loss = get_results(pred_family, real_family)
    genus_score, genus_loss = get_results(pred_genus, real_genus)
    species_score, species_loss = get_results(pred_species, real_species)
    ham_loss_final.append((family_loss + genus_loss + species_loss)/3)
    ham_score_final.append((family_score + genus_score + species_score)/3)
    dist = (family_loss + genus_loss + species_loss)/3 * y.shape[0]
    ham_dist_final.append(dist)
    optimal.append(best_k)
    majority_label.append(final_label)
    print("Hamming loss:", ham_loss_final[-1])
    print("Hamming score:", ham_score_final[-1])
    print("Hamming distance:", ham_dist_final[-1])
    print("=================================================================================")
    
#     print(ham_loss_final, ham_score_final, ham_dist_final)
```

```python
count = 0
col = ['Iteration No.','Cluster No.','Majority Class-Family','Majority Class-Species','Majority Class-Genus']
final_data1 = pd.DataFrame(columns = col)
for i in range(len(majority_label)):
    for j in range(best_k):
        iteration = "Iteration-" + str(i)
        final_data1.at[count, 'Iteration No.'] = iteration
        cluster = 'Cluster-' + str(j)
        final_data1.at[count, 'Cluster No.'] = cluster
        final_data1.at[count, 'Majority Class-Family'] = majority_label[i][j]['Family']
        final_data1.at[count, 'Majority Class-Genus'] = majority_label[i][j]['Genus']
        final_data1.at[count, 'Majority Class-Species'] = majority_label[i][j]['Species']
        count = count + 1
```
